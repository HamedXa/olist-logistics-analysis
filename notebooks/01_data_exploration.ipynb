```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
sellers = pd.read_csv("olist_sellers_dataset.csv")
```

```python


# Cell 3 - Convert date columns
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])

# Cell 4 - Filter to delivered orders only
delivered = orders[orders["order_status"] == "delivered"].copy()
print(f"Delivered orders: {len(delivered):,}")

# Cell 5 - Engineer delivery KPIs
delivered["delivery_time_days"] = (
    delivered["order_delivered_customer_date"] - 
    delivered["order_purchase_timestamp"]
).dt.days

delivered["estimated_days"] = (
    delivered["order_estimated_delivery_date"] - 
    delivered["order_purchase_timestamp"]
).dt.days

delivered["delay_days"] = (
    delivered["order_delivered_customer_date"] - 
    delivered["order_estimated_delivery_date"]
).dt.days

delivered["is_late"] = delivered["delay_days"] > 0

# Cell 6 - Summary stats
print(f"\nAverage delivery time: {delivered['delivery_time_days'].mean():.1f} days")
print(f"Late deliveries: {delivered['is_late'].sum():,} ({delivered['is_late'].mean()*100:.1f}%)")
print(f"On-time rate: {(~delivered['is_late']).mean()*100:.1f}%")

# Cell 7 - Save cleaned file
delivered.to_csv("delivered_orders_clean.csv", index=False)
print("\nCleaned file saved.")
```
